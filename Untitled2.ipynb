```python
from pandas import *
import os.path
from scipy import interpolate 
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder = os.listdir('/content/drive/MyDrive/Group2Module2_CSS844/processed_files')
path = '/content/drive/MyDrive/Group2Module2_CSS844/processed_files/'

all_xs = []
all_ys = []
all_zs = []

for arr in folder:
  #filepath = os.path.join(folder, arr)
  dataloop = np.load(path + arr)
  x = dataloop[:, 0]
  y = dataloop[:, 1]
  z = dataloop[:, 2]
  all_xs.extend(x)
  all_ys.extend(y)
  all_zs.extend(z)

all_xs = np.array(all_xs)
all_ys = np.array(all_ys)
all_zs = np.array(all_zs)


# setting the size of figure shown
fig = plt.figure(figsize=(8, 8))
# type of figure (3D)
ax = fig.add_subplot(111, projection='3d')

# Plot scatter plot
ax.scatter(all_xs, all_ys, all_zs)
plt.show()


ax2 = plt.axes(projection='3d')
ax2.plot_surface(all_xs, all_ys, all_zs, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax2.set_title('TITLE')

ax3 = plt.axes(projection='3d')
ax3.scatter(all_xs, all_ys, all_zs, c=all_zs, cmap='viridis', linewidth=0.5)
```
